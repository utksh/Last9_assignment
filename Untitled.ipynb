```python
import datetime

dt = datetime.datetime(2019, 3, 15,18)
end = datetime.datetime(2021, 3, 28,18)
step = datetime.timedelta(hours=24)

result = []

while dt < end:
    result.append(dt.strftime('%Y-%m-%d %H:%M'))
    dt += step
```

```python
initial_reading = 10
weekend = [5,6]
summer_months = [4,5,6,7,8]
summer_pers = np.arange(10,20)
winter_pers = np.arange(8,18)
consumption_dict = {}
for date in result:
    stripped = datetime.datetime.strptime(date, '%Y-%m-%d %H:%M')
    day = stripped.weekday()
    month = stripped.month
    if month in summer_months:
        if day in weekend:
            # Make each of the last 5 elements 5x more likely
            prob = [1.0]*(len(summer_pers)-5) + [5.0]*5
            
            # Normalising to 1.0
            prob /= np.sum(prob)
            day_consumption  = np.random.choice(summer_pers, p = prob)
        else:
            
            day_consumption  = np.random.choice(summer_pers)
    else:
        if day in weekend:
            # Make each of the last 5 elements 5x more likely
            prob = [1.0]*(len(winter_pers)-5) + [5.0]*5
            
            # Normalising to 1.0
            prob /= np.sum(prob)
            day_consumption  = np.random.choice(winter_pers, p = prob)
       
        else:
            
            day_consumption  = np.random.choice(winter_pers)
            
        
    initial_reading += day_consumption
    
    
    consumption_dict[date] = initial_reading
            
print()
```

```python
noise = np.random.laplace(loc=0, scale=1, size=len(consumption_dict)) 
```

```python
"""Brutlag Algorithm"""
PERIOD = 12        # The given time series has seasonal_period=12
GAMMA = 0.3684211  # the seasonility component
SF = 1.96          # brutlag scaling factor for the confidence bands.
UB = []            # upper bound or upper confidence band
LB = []            # lower bound or lower confidence band
```

```python
model = ExponentialSmoothing(
    df_train, trend='additive', seasonal='additive').fit()
prediction = model.predict(
    start= df.iloc[:, 0].index[0], end=df.iloc[:, 0].index[-1])
```

```python
difference_array = []
dt = []
difference_table = {
    "actual": df.iloc[:, 0], "predicted": prediction, "difference": difference_array, "UB": UB, "LB": LB}
```

```python
"""Calculatation of confidence bands using brutlag algorithm"""
for i in range(len(prediction)):
    diff = df.iloc[:, 0][i]-prediction[i]
    if i < PERIOD:
        dt.append(GAMMA*abs(diff))
    else:
        dt.append(GAMMA*abs(diff) + (1-GAMMA)*dt[i-PERIOD])

    difference_array.append(diff)
    UB.append(prediction[i]+SF*dt[i])
    LB.append(prediction[i]-SF*dt[i])

print("\nDifference between actual and predicted\n")
difference = pd.DataFrame(difference_table)
print(difference)
```
